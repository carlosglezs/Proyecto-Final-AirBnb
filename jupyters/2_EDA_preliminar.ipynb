# Análisis preliminar del conjunto de datos

```python
# Importaciones
import pandas as pd
import numpy as np
import sys
import ast
sys.path.append("..")
from src import soporte as sp
from datetime import datetime
```

Se cargan los archivos con las ofertas de Airbnb de los 3 trimestres

```python
# Leer archivos
df_raw_mar = pd.read_csv("../data/airbnb_mad_mar.csv")
df_raw_dec = pd.read_csv("../data/airbnb_mad_dec.csv")
df_raw_sep = pd.read_csv("../data/airbnb_mad_sep.csv")
```

```python
#Generamos copia de seguridad
df_mar = df_raw_mar.copy()
df_dec = df_raw_dec.copy()
df_sep = df_raw_sep.copy()
```

```python
#Fusionamos los datos uno debajo de otro
df_year = pd.concat([ df_mar, df_dec, df_sep], ignore_index=True)
```

Realizamos un EDA preeliminar de todo el archivo unido

```python
sp.eda_preliminar(df_year)
```

Seleccionamos las columnas con las que vamos a trabajar

```python
df_rows  = df_year[['id', 'review_scores_rating', 'last_scraped', 'neighbourhood_cleansed', 'property_type', 'accommodates', 'bathrooms','bedrooms',	'beds','amenities'	, 'price', 'minimum_nights','number_of_reviews',	 'review_scores_accuracy',	'review_scores_cleanliness',	'review_scores_checkin',	'review_scores_communication', 'review_scores_location',	'review_scores_value' ]]
```

```python
# Convertir la columna 'fecha_reserva' a tipo datetime
df_rows['last_scraped'] = pd.to_datetime(df_year['last_scraped'], format='%Y-%m-%d')
```

Realizamos un EDA preeliminar de las columnas que nos interesan

```python
sp.eda_preliminar(df_rows)
```

Generamos la lista de los servicios que vamos a analizar

```python
services = [
    "Kitchen", "TV", "Air conditioning", "Pets allowed", "Dryer", "Patio or balcony",
    "Iron", "Microwave", "Cooking basics",  "Self check-in",
    "Washer", "Heating", "Hair dryer", "Wifi", "Coffee maker",  "Elevator",'Pool'
]
```

```python
# Convertir strings en listas de verdad
df_rows['amenities'] = df_rows['amenities'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])
```

```python
for service in services:
    df_rows[service] = df_rows['amenities'].apply(lambda x: service in x)
```

```python
#Convertimos todos los datos en minusculas
sp.valores_minus(df_rows)
```

```python
# Convertimos todos los nombres de las columnas a minusculas y se sustituye los espacios por "-"
df_rows.columns = df_rows.columns.str.lower().str.replace(' ', '_')

```

```python
# Obsertvamos posibles duplicados
df_rows.duplicated().sum()
```

```python
# Eliminamos el simbolo del $ y lo convertimos a float 
df_rows['price'] = df_rows['price'].replace('[\$,]', '', regex=True).astype(float)
```

Generamos una columna con el trimestre al que pertenece cada valor para poder unirlo con el dataset de las reservas

```python
# Crear una columna de trimestre
df_rows['trimester'] = df_rows['last_scraped'].dt.to_period('Q')
```

```python
# Eliminamos las columna amenities ya que no nos sirve
df_rows = df_rows.drop(columns=['amenities'])
```

```python
# Consultamos si hay duplicados
duplicados = df_rows[df_rows.duplicated(subset=['id', 'trimester'], keep=False)]
```

Cargamos el dataset de las reservas para fusionarlo con el de las ofertas

```python
df_resumen = pd.read_csv("../data/resumen_anual.csv")
```

Lo unimos en base a coincidencia por 'id' y 'trimester'

```python
df_rows['trimester'] = df_rows['trimester'].astype(str)
df_resumen['trimester'] = df_resumen['trimester'].astype(str)

df_merged = pd.merge(
    df_rows,
    df_resumen,
    left_on=['id', 'trimester'],
    right_on=['listing_id', 'trimester'],
    how='inner'
)
```

```python
# Borramos la columna 'listing_id' sobrante
df_merged = df_merged.drop(columns=['listing_id'])
```

```python
#Volvemos a dar formato temporal a trimester
df_merged['trimester'] = pd.PeriodIndex(df_merged['trimester'], freq='Q')
```

```python
#Eliminamos columnas no deseadas que empezaban por 'unnamed'
df_merged = df_merged.loc[:, ~df_merged.columns.str.contains('^Unnamed')]
```

```python
#Creamos columna para análizar el precio por persona
df_merged['price_person'] = df_merged['price'] / df_merged['accommodates']
```

```python
df_merged.to_csv('../data/data_anual.csv', index=False, )
```
